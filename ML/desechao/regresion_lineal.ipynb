# Regresión Lineal para Predicción de Humedad/Temperatura
Este notebook implementa un modelo de regresión lineal para predecir valores continuos como humedad o temperatura, replicando el análisis del notebook original.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Cargar el dataset (ajusta la ruta si es necesario)
dataset = pd.read_csv('../dataset/ML1_temperature.xlsx - Hoja1.csv', decimal=',')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1:].values
```

```python
# Separar en entrenamiento y validación
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Entrenar modelo de regresión lineal
model = LinearRegression()
model.fit(x_train, y_train)
y_train_pred = model.predict(x_train)
y_val_pred = model.predict(x_val)
```

```python
# Evaluar desempeño
train_loss = mean_squared_error(y_train, y_train_pred)
val_loss = mean_squared_error(y_val, y_val_pred)
print(f'Training Loss (MSE): {train_loss}')
print(f'Validation Loss (MSE): {val_loss}')
```

```python
# Graficar resultados
plt.figure(figsize=(12, 6))
plt.scatter(y_train, y_train_pred, alpha=0.5, label='Entrenamiento')
plt.scatter(y_val, y_val_pred, alpha=0.5, label='Validación')
plt.xlabel('Valor Real')
plt.ylabel('Predicción')
plt.title('Regresión Lineal: Real vs Predicho')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Explicación breve
print('La regresión lineal es un modelo simple y fácil de interpretar. El compromiso sesgo-varianza se observa comparando el error de entrenamiento y validación. Si ambos errores son altos, hay alto sesgo; si hay mucha diferencia, hay alta varianza.')
```
